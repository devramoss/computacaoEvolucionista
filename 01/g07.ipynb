```python
import random
import matplotlib.pyplot as plt
import numpy as np
import math
import statistics
```

```python
def f(x1, x2, x3, x4, x5, x6, x7, x8, x9, x10):
    return math.pow(x1, 2) + math.pow(x2, 2) + (x1*x2) - (14*x1) - (16*x2) + (math.pow(x3-10, 2)) + 4*(math.pow(x4-5, 2)) + (math.pow(x5-3, 2)) + 2*(math.pow(x6-1, 2)) + 5*(math.pow(x7, 2)) + 7*(math.pow(x8-11, 2)) + 2*(math.pow(x9-10, 2)) + (math.pow(x10-7, 2)) + 45
```

```python
def factible(x1, x2, x3, x4, x5, x6, x7, x8, x9, x10):
    g = [None, None, None, None, None, None, None, None]
    g[0] = -105 + (4*x1) + (5*x2) - (3*x7) + (9*x8)
    g[1] = (10*x1) - (8*x2) - (17*x7) + (2*x8)
    g[2] = (-8*x1) + (2*x2) + (5*x9) - (2*x10) - 12
    g[3] = 3*math.pow(x1-2, 2) + 4*math.pow(x2-3, 2) + 2*math.pow(x3, 2) - (7*x4) -120
    g[4] = 5*math.pow(x1, 2) + (8*x2) + math.pow(x3-6, 2) - (2*x4) - 40
    g[5] = math.pow(x1, 2) + 2*math.pow(x2-2, 2) - (2*x1*x2) + (14*x5) - (6*x6)
    g[6] = 0.5*math.pow(x1-8, 2) + 2*math.pow(x2-4, 2) + 3*math.pow(x5, 2) - x6 -30
    g[7] = -(3*x1) + (6*x2) + 12*math.pow(x9-8, 2) - (7*x10)

    return g
```

```python
def busca_aleatoria(quantidade_iteracoes, limite_inferior, limite_superior):
    melhor_w=np.inf

    nFactivel=0

    melhor_x1=None
    melhor_x2=None
    melhor_x3=None
    melhor_x4=None
    melhor_x5=None
    melhor_x6=None
    melhor_x7=None
    melhor_x8=None
    melhor_x9=None
    melhor_x10=None

    melhores_dominios_x1=[]
    melhores_dominios_x2=[]
    melhores_dominios_x3=[]
    melhores_dominios_x4=[]
    melhores_dominios_x5=[]
    melhores_dominios_x6=[]
    melhores_dominios_x7=[]
    melhores_dominios_x8=[]
    melhores_dominios_x9=[]
    melhores_dominios_x10=[]

    melhores_imagens =[]

    dominio_x1=[]
    dominio_x2=[]
    dominio_x3=[]
    dominio_x4=[]
    dominio_x5=[]
    dominio_x6=[]
    dominio_x7=[]
    dominio_x8=[]
    dominio_x9=[]
    dominio_x10=[]

    imagem=[]

    for _ in range(quantidade_iteracoes):
        X1 = random.uniform(limite_inferior[0], limite_superior[0])
        X2 = random.uniform(limite_inferior[1], limite_superior[1])
        X3 = random.uniform(limite_inferior[2], limite_superior[2])
        X4 = random.uniform(limite_inferior[3], limite_superior[3])
        X5 = random.uniform(limite_inferior[4], limite_superior[4])
        X6 = random.uniform(limite_inferior[5], limite_superior[5])
        X7 = random.uniform(limite_inferior[6], limite_superior[6])
        X8 = random.uniform(limite_inferior[7], limite_superior[7])
        X9 = random.uniform(limite_inferior[8], limite_superior[8])
        X10= random.uniform(limite_inferior[9], limite_superior[9])

        w = f(X1, X2, X3, X4, X5, X6, X7, X8, X9, X10)


        if melhor_w > w and all(i >= 0 for i in factible(X1, X2, X3, X4, X5, X6, X7, X8, X9, X10)):
            melhor_w = w

            nFactivel+=1

            melhor_x1 = X1
            melhor_x2 = X2
            melhor_x3 = X3
            melhor_x4 = X4
            melhor_x5 = X5
            melhor_x6 = X6
            melhor_x7 = X7
            melhor_x8 = X8
            melhor_x9 = X9
            melhor_x10= X10

        else:
          w+=2*24.000

        melhores_dominios_x1.append(melhor_x1)
        melhores_dominios_x2.append(melhor_x2)
        melhores_dominios_x3.append(melhor_x3)
        melhores_dominios_x4.append(melhor_x4)
        melhores_dominios_x5.append(melhor_x5)
        melhores_dominios_x6.append(melhor_x6)
        melhores_dominios_x7.append(melhor_x7)
        melhores_dominios_x8.append(melhor_x8)
        melhores_dominios_x9.append(melhor_x9)
        melhores_dominios_x10.append(melhor_x10)

        melhores_imagens.append(melhor_w)

        dominio_x1.append(X1)
        dominio_x2.append(X2)
        dominio_x3.append(X3)
        dominio_x4.append(X4)
        dominio_x5.append(X5)
        dominio_x6.append(X6)
        dominio_x7.append(X7)
        dominio_x8.append(X8)
        dominio_x9.append(X9)
        dominio_x10.append(X10)

        imagem.append(w)

    return melhor_x1, melhor_x2, melhor_x3, melhor_x4, melhor_x5, melhor_x6, melhor_x7, melhor_x8, melhor_x9, melhor_x10, melhor_w, nFactivel, melhores_imagens
```

```python
# Definindo os limites
limites_inferior = [-10, -10, -10, -10, -10, -10, -10, -10, -10, -10]
limites_superior = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10]

todas_imagens = []

for i in range(10):
  # Gerando uma nova semente a cada iteração
  random.seed(i)

  print("\nExecução: ",  i)

  melhor_x1, melhor_x2, melhor_x3, melhor_x4, melhor_x5, melhor_x6, melhor_x7, melhor_x8, melhor_x9, melhor_x10, melhor_w, nFactivel, melhores_imagens = busca_aleatoria(8000, limites_inferior, limites_superior)

  print("\nSoluções factíveis: ", nFactivel, "\n\nMelhor x1: ", melhor_x1, "\nMelhor x2: ", melhor_x2, "\nMelhor x3: ", melhor_x3, "\nMelhor x4: ", melhor_x4, "\nMelhor x5: ", melhor_x5, "\nMelhor x6: ", melhor_x6, "\nMelhor x7: ", melhor_x7, "\nMelhor x8: ", melhor_x8, "\nMelhor x9: ", melhor_x9, "\nMelhor x10: ", melhor_x10, "\nMelhor w: ", melhor_w)

  todas_imagens.append(melhores_imagens)

  #visualizando os resultados
  plt.title('\nConvergência')
  plt.xlabel('\nIterações')
  plt.ylabel('\nMelhor valor')
  plt.scatter(range(len(melhores_imagens)), melhores_imagens,color='r')
  plt.grid(True)
  plt.show()

  # Remover valores infinitos e NaN da lista melhores_imagens
  melhores_imagens = [x for x in melhores_imagens if x != float('inf') and not math.isnan(x)]

  # Calculando desvio padrão se houver mais de um elemento em melhores_imagens
  if len(melhores_imagens) > 1:
      mediana = statistics.median(melhores_imagens)
      media = statistics.mean(melhores_imagens)
      desvio_padrao = statistics.stdev(melhores_imagens)
  else:
      desvio_padrao = 0
      media = 0
      mediana = 0

  print("\nMediana: ", mediana)
  print("\nMédia: ", media)
  print("\nDesvio padrão: ", desvio_padrao)

#visualizando todas as convergências em um único gráfico
plt.title('\nConvergências')
plt.xlabel('\nIterações')
plt.ylabel('\nMelhor valor')

for i, melhores_imagens in enumerate(todas_imagens):
    plt.scatter(range(len(melhores_imagens)), melhores_imagens, label=f'Execução {i}', alpha=0.5)

plt.legend()
plt.grid(True)
plt.show()
```
